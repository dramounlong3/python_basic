```python
df = pd.DataFrame(columns=['age','score'])
df['score'] = np.random.randint(100, size=(10))
df['age'] = np.random.randint(25, size=(10))
```

```python
df
```

```python
df.plot(x = 'age', y = 'score', kind='scatter')
```

```python
sorted_df = df.sort_values(by=['age'])
sorted_df
```

```python
sorted_df.plot(x='age',y='score')
```
